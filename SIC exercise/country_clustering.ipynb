```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('/content/drive/MyDrive/SIC/0.Samsung CSR AIML/Resources/SIC-Artificial-Intelligence/SIC_AI_Coding_Exercises/SIC_AI_Chapter_05_Coding_Exercises/Clustering_Country.csv')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
cluster_df = df.drop('country', axis=1)

cluster_df.head()
```

```python
cluster = KMeans(n_clusters=3, random_state=0)

cluster.fit(cluster_df)
```

```python
cluster.fit_predict(cluster_df)
```

```python
df['cluster'] = cluster.fit_predict(cluster_df)

df.head()
```

```python
df['cluster'].value_counts()
```

```python
df['cluster'].value_counts().hist()
```

```python
import seaborn as sns

sns.pairplot(df, hue='cluster')
```
